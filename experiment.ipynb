```python
def count_params(model):
    count = sum(p.numel() for p in model.parameters() if p.requires_grad)
    # turn into string, then insert commas every 3 digits from the right
    count_str = f"{count:,}"
    return count_str
```

```python
from model import build_model

# backbone = "resnet101"
backbone = "vgg16"

unet1 = build_model(model_name="Unet", backbone=backbone)
print(f"Number of parameters in the model Unet: {count_params(unet1)}")

unet2 = build_model(model_name="UnetPlusPlus", backbone=backbone)
print(f"Number of parameters in the model UnetPlusPlus: {count_params(unet2)}")

unet3 = build_model(model_name="Unet3Plus", backbone=backbone)
print(f"Number of parameters in the model Unet3Plus: {count_params(unet3)}")
```
